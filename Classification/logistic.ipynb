```python
import os
import pandas as pd
'''
def load_data(path):
    X_train_data_path = os.path.join(path, "X_train.csv")
    Y_train_data_path = os.path.join(path, "Y_train.csv")
    X_test_data_path = os.path.join(path, "X_test.csv")
    return (pd.read_csv(X_train_data_path),
            pd.read_csv(Y_train_data_path),
            pd.read_csv(X_test_data_path))
    '''
    
def read_data(path):
    train_data_path = os.path.join(path, "X_train.csv")
    train_label_path = os.path.join(path, "Y_train.csv")
    test_data_path = os.path.join(path, "X_test.csv")
    f = open(train_data_path)
    X_train = pd.read_csv(f, sep=',', header=0)
    f.close()

    #X_train = np.array(X_train.values)
    f = open(train_label_path)
    Y_train = pd.read_csv(f, sep=',', header=0)
    f.close()

    #Y_train = np.array(Y_train.values)
    f = open(test_data_path)
    X_test = pd.read_csv(f, sep=',', header=0)
    f.close()

    #X_test = np.array(X_test.values)
    return (X_train, Y_train, X_test) 
   
X_train, Y_train, X_test = read_data("E:\郭志\ML\Classification\Data")
```

```python
import numpy as np

X_train = np.array(X_train.values)
Y_train = np.array(Y_train.values)
X_test = np.array(X_test.values)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
import matplotlib.pyplot as plt
from sklearn import preprocessing

model = LogisticRegression(penalty='l2') 
plt.scatter(X_train[:, 0], X_train[:, 1], c=Y_train)
plt.show()
X = preprocessing.scale(X_train)    # normalization step
#X=X_train
plt.scatter(X[:, 0], X[:, 1], c=Y_train)
plt.show()
train_x, test_x, train_y, test_y = train_test_split(X, Y_train, test_size=.3)
model.fit(train_x, train_y)  
```

```python
print(model.score(test_x, test_y))
```
